# A Harder Challenge: A Shifting Environment

In this tutorial we will tackle the Cart-Pole task in the Open AI gym.  This task is harder because it adds two new elements.  The state space is continuous and has a temporal element as well.  Therefore our agent must take *observations* from the environment. Second, the reward (or rather lack of punishment) is delayed.  A mistake may lead to failure, but not necessarily in the next frame. 

One consequence of this is that we will need to introduce a new concept to our model.  Memory!  Our agents memory will allow us to store the agents experiences and use them as a training set.  Importantly, it will allow us to learn on more than one example at a time!

We are deviating from [Arthur's implementation](https://github.com/awjuliani/DeepRL-Agents/blob/master/Vanilla-Policy.ipynb) this time.

```python
import gym
import tensorflow as tf
import numpy as np

env = gym.make('CartPole-v0')
```

```python
# Run to get an idea of the state and rewards.  
# The render is fun to view as well if not being run in Colab.
env.reset()
rewards = []
for _ in range(50):
    # If not in Colab, else comment out
    env.render()
    state, reward, done, info = env.step(env.action_space.sample()) # take a random action
    print(state, reward, done, info)
    rewards.append(reward)
    if done:
        print(f'Lost in {len(rewards)} steps')
        rewards = []
        env.reset()
```

We can see that we get the state of the environment in terms of 4 real numbers.  Our goal as usual will be to make a prediction about which action to take (left or right for this game).  Let's construct the the Q-Network as we have done before.

```python
class QLearner:
    def __init__(self, learning_rate=0.01, state_size=4, action_size=2, hidden_size=10, name='QNetwork'):
        with tf.variable_scope(name):
            with tf.variable_scope('forwards'):
                # Inputs are given by a 4-col vector with some unknown number of examples
                self.inputs_ = tf.placeholder(tf.float32, [None, state_size], name='inputs')
                # Are we in training mode?
                self.train_mode = tf.placeholder(tf.bool)                
                
                self.fc1 = tf.layers.dense(self.inputs_, hidden_size, activation=tf.nn.relu)
                #self.norm = tf.contrib.layers.layer_norm(self.fc1)
                #self.drop = tf.layers.dropout(self.fc1, training=self.train_mode)
                self.fc2 = tf.layers.dense(self.fc1, hidden_size, activation=tf.nn.relu)
                
                self.output = tf.layers.dense(self.fc2, action_size, activation=None)
                self.chosen_action = tf.distributions.Categorical(probs=self.output).sample(1)[0]
                
            with tf.variable_scope('backwards'):
                # Actions are given as an integer
                self.actions_ = tf.placeholder(tf.int32, [None], name='actions')
                one_hot_actions = tf.one_hot(self.actions_, action_size)
            
                # Target Values from the Bellman Equation
                self.targetQs_ = tf.placeholder(tf.float32, [None], name='target')
                
                # Train with loss (targetQ - Q)^2
                self.Q = tf.reduce_sum(tf.multiply(self.output, one_hot_actions), axis=1)
                
                self.loss = tf.reduce_mean(tf.square(self.targetQs_ - self.Q))
                self.opt = tf.train.AdamOptimizer(learning_rate).minimize(self.loss)
```

## Adding Memory

We will implement memory as a double-ended que.  This acts nicely as a buffer because as it gets to a maximum size, it will jetison the oldest experiences.  This lets us maintain a manageable size of training experiences which prioritizing the most recent (and hopefully useful) ones. Each experience consists of an initial state, action, reward, and the next state. 

```python
from collections import deque
class Memory():
    def __init__(self, max_size=1000):
        self.buffer = deque(maxlen=max_size)
        
    def add(self, experience):
        """Add experiences to the front end of our memory
        
        Will push out oldest memories once at max_size automatically
        """
        self.buffer.append(experience)
        
    def sample(self, batch_size):
        """Sample random examples from our Memory"""
        idx = np.random.choice(np.arange(len(self.buffer)),
                               size = batch_size,
                               replace=False)
        return [self.buffer[ii] for ii in idx]
```

```python
# Hyperparameters

## Learning Parameters
train_episodes = 1500          # max number of episodes to learn from
max_steps = 200                # max steps in an episode
gamma = 0.99                   # future reward discount

## Exploration parameters
explore_start = 1.0            # exploration probability at start
explore_stop = 0.01            # minimum exploration probability 
decay_rate = 0.0001            # exponential decay rate for exploration prob

## Network parameters
hidden_size = 16                # number of units in each Q-network hidden layer
learning_rate = 0.00005         # Q-network learning rate

## Memory parameters
memory_size = 10000            # memory capacity
batch_size = 128               # experience mini-batch size
pretrain_length = batch_size   # number experiences to pretrain the memory
```

```python
tf.reset_default_graph()
mainQN = QLearner(hidden_size=hidden_size, learning_rate=learning_rate)
```

## The Cold Start Problem

When we start out we don't yet have any data to train our model on.  A nice way to solve that problem is to generate a dataset through random play.  We'll start by generating a bunch of experiences which are entirely random.  This will be enough to get our model started.

```python
# Initialize the Simulation
env.reset()

# Take one random step to get the pole and cart moving
state, reward, done, _ = env.step(env.action_space.sample())

memory = Memory(max_size=memory_size)

# Make a bunch of random actions and store the experiences
for _ in range(pretrain_length):
    #env.render()
    
    # Make a random action
    action = env.action_space.sample()
    next_state, reward, done, _ = env.step(action)
    
    if done:
        # We place a filler of zeros in for the next state
        next_state = np.zeros(state.shape)
        # Add complete experience to memory
        memory.add((state, action, reward, next_state))
        
        #Start new episode
        env.reset()
        # Take on random step to get the pole and cart moving
        state, reward, done, _ = env.step(env.action_space.sample())
        
    else:
        # Add experience to memory
        memory.add((state, action, reward, next_state))
        state = next_state
```

## Time to Train

We will now go through the training process.  The training process is really only cosmetically different from what we've done before.  

```python
saver = tf.train.Saver()
rewards_list = []
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    #Start new episode
    env.reset()
    # Take on random step to get the pole and cart moving
    state, reward, done, _ = env.step(env.action_space.sample())
    step = 0 
    for ep in range(1, train_episodes):
        total_reward = 0
        t = 0
        while t < max_steps:
            step += 1
            #env.render()
            
            explore_p = explore_stop + (explore_start - explore_stop)*np.exp(-decay_rate*step)
            if explore_p < np.random.rand():
                # Make a random action
                action = env.action_space.sample()
            else:
                # Get action from Q-Network
                Qs, action = sess.run([mainQN.output, mainQN.chosen_action], feed_dict={mainQN.inputs_: state.reshape((1, *state.shape)), mainQN.train_mode: False})
                action = action[0]
            
            # Take Action, Get Reward
            next_state, reward, done, _ = env.step(action)
            
            total_reward += reward
            
            if done:
                # Fill in the next_state with zeros
                next_state = np.zeros(state.shape)
                t = max_steps
                
                if ep % 50 == 0:
                    print(f'Episode: {ep}',
                          f'Reward: {total_reward}',
                          f'Training loss: {loss}',
                          f'Explore P: {explore_p}')
                rewards_list.append((ep, total_reward))
                
                # Add experience to memory
                memory.add((state, action, reward, next_state))
                
                #Start new episode
                env.reset()
                # Take on random step to get the pole and cart moving
                state, reward, done, _ = env.step(env.action_space.sample())
            
            else:
                # Add experience to memory
                memory.add((state, action, reward, next_state))
                state = next_state
                t += 1
            
            # Training Time!
            # Get mini-batch
            batch = memory.sample(batch_size)
            states = np.array([each[0] for each in batch])
            actions = np.array([each[1] for each in batch])
            rewards = np.array([each[2] for each in batch])
            next_states = np.array([each[3] for each in batch])
            
            # Train
            target_Qs = sess.run(mainQN.output, feed_dict={mainQN.inputs_: next_states, mainQN.train_mode: False})
            
            # Set target Qs to 0 for states where the episode ends
            episode_ends = (next_states == np.zeros(states[0].shape)).all(axis=1)
            target_Qs[episode_ends] = (0, 0)
            
            # Set targets per the Bellman Equation
            targets = rewards + gamma * np.max(target_Qs, axis=1)
            
            loss, _ = sess.run([mainQN.loss, mainQN.opt],
                                feed_dict={mainQN.inputs_: states,
                                           mainQN.targetQs_: targets,
                                           mainQN.actions_: actions,
                                           mainQN.train_mode: True})
            
    saver.save(sess, "checkpoints/cartpole.ckpt")
```
